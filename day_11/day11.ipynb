```python
import math
data = [instruction.replace("\n","") for instruction in open('input.txt').readlines()]

class Monkey:
    def __init__(self, number,items,operation,test,trueCase,falseCase):
        self.number = number
        self.items = items
        self.operation = operation
        self.test = test
        self.trueCase = trueCase
        self.falseCase = falseCase

monkeys = []       
lcm = 0
def initMonkeys():
    monkeys.clear()
    for i in range(0,len(data),7):
        number = data[i][-2:-1]
        items = [int(num) for num in data[i+1][18:].split(",")]
        test = int(data[i+3][21:])
        op = lambda old,func=data[i+2][19:] : eval(func)
        trueCase = int(data[i+4][-1:])
        falseCase = int(data[i+5][-1:])
        monkeys.append(Monkey(number,items,op,test,trueCase,falseCase))

    lcm = monkeys[0].test
    for monkey in monkeys[1:]:
        lcm = lcm*monkey.test//math.gcd(lcm,monkey.test)
    
    return lcm
```

```python
from math import floor

def printState():
    print("The monkeys are holding items with these worry levels:")
    for monkey in monkeys:
        print("Monkey " + monkey.number + ": " + str(monkey.items))
    print("\n")
    
def p1():
    initMonkeys()
    counts = [0]*8
    for turn in range(0,20):
        for monkey in monkeys:
            for item in monkey.items:
                counts[int(monkey.number)] += 1
                if (floor(monkey.operation(item)/3) % monkey.test) == 0:
                    monkeys[monkey.trueCase].items.append(floor(monkey.operation(item)/3))
                else:
                    monkeys[monkey.falseCase].items.append(floor(monkey.operation(item)/3))
            monkey.items.clear()
    printState()
    counts.sort()
    return (counts[-1]*counts[-2])           
        

print(p1())
```

```python
def printState():
    print("The monkeys are holding items with these worry levels:")
    for monkey in monkeys:
        print("Monkey " + monkey.number + ": " + str(monkey.items))
    print("\n")
    
def printCounts(counts,round):
    print("== After round " + str(round+1) + " ==")
    mk = 0
    for count in counts:
        print("Monkey " + str(mk) + ": " + str(counts[mk]))
        mk += 1
    print("\n")
    
def p2():
    lcm = initMonkeys()
    counts = [0]*len(monkeys)
    for turn in range(0,10000):
        for monkey in monkeys:
            for item in monkey.items:
                counts[int(monkey.number)] += 1
                num = monkey.operation(item) % lcm
                if (num % monkey.test) == 0:
                    monkeys[monkey.trueCase].items.append(num)
                else:
                    monkeys[monkey.falseCase].items.append(num)
            monkey.items.clear()
        if turn % 1000 == 0:  
            printCounts(counts,turn)
    printCounts(counts,10000)
    counts.sort()
    return (counts[-1]*counts[-2])           
        

print(p2())
```
