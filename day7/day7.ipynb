```python
data = [num.replace("\n","") for num in open('input.txt').readlines()]
unitTest = [num.replace("\n","") for num in open('unittest.txt').readlines()]
```

```python
import re
from anytree import Node, RenderTree, PreOrderIter,LevelOrderIter, PostOrderIter

def makeTree(io):
    root = Node("root",parent=None,size=0)
    curr = root

    for command in io[1:]:
        if '$ cd' in command:
            if '..' in command:
                curr = curr.parent
            else:
                curr = Node(command[5:],parent=curr,size=0)
        elif 'dir' not in command and '$ ls' not in command:
            curr.size += int(re.findall(r'\d+', command)[0])
        
    for node in PostOrderIter(root):
        if node.name == 'root':
            continue
        node.parent.size += node.size

    return root

def printFS(root):
    for row in RenderTree(root):
        print("%s%s (%ib)" % (row.pre, row.node.name,row.node.size))

def P1(root):
    p1 = 0
    for node in PreOrderIter(root):
        if node.size < 100000:
            p1 += node.size
        
    return p1

    
def P2(root):
    spaceNeeded = root.size-40000000
    return min([node.size for node in LevelOrderIter(root, stop=lambda n: n.size < spaceNeeded)])
    
#printFS(makeTree(unitTest))
print("Answer for Part1: " + str(P1(makeTree(data)))+"\n")
print("Answer for Part1: " + str(P2(makeTree(data)))+"\n")
printFS(makeTree(data))
```
